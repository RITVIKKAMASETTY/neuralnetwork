```python
from tensorflow.keras.models import load_model

# Load the existing model
model = load_model("accident_detection_model.h5", compile=False)  # Load without compiling

# Compile the model properly
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

# Save again with optimizer included
model.save("accident_detection_model.h5", include_optimizer=True)
```
